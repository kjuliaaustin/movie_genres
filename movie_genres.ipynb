```python
#import libraries
import pandas as pd
from pandas import Series, DataFrame
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df.info()
```

```python
df.drop_duplicates(inplace = True)
```

```python
#Matching id count with genre count
df.dropna(subset = ['genres'], inplace = True)
```

#Data Preparation

```python
df.head()
```

```python
#Add a profit column by subtracting the budget from the revenue.
df['profit'] = df['revenue'] - df['budget']
```

```python
#Only show columns that we need.
movies_genre = df[['popularity','budget','revenue','original_title','runtime','genres','release_date','vote_count','vote_average','profit']]
movies_genre.head()
```

```python
#Splitting the genres up per data entry
split = movies_genre['genres'].str.split('|').apply(Series,1).stack()
split.index = split.index.droplevel(-1)
split.name = 'genres_split'
del movies_genre['genres']
movies_genre = movies_genre.join(split)

movies_genre
```

#Research Hypotheses

```python
#Get rid of split genres
df = pd.read_csv('imdb_movies.csv')
df.drop_duplicates(inplace = True)
df['profit'] = df['revenue'] - df['budget']
movies_genre = df[['popularity','budget','revenue','original_title','runtime','genres','release_date','vote_count','vote_average','profit']]
movies_genre.head()
```

**Research Hypothesis (H):**

1. The best movies according to vote avg. return high profit and revenue.

```python
#Counting movies only with a vote count above 50 and excluding strings
movies_counted = movies_genre[movies_genre['vote_count'] >= 50].select_dtypes(include=['int64', 'float64'])

movies_counted.corr(method = 'spearman')
```

```python
#Plot the correlation between profit and vote average
sns.regplot(x = 'vote_average', y = 'profit', data = movies_counted, line_kws = {'color': 'red'})
```

```python
#Plot the correlation between revenue and vote average
sns.regplot(x = 'vote_average', y = 'revenue', data = movies_counted, line_kws = {'color': 'red'})
```

**Research Hypothesis (H):**

2. The best movies according to popularity return high profit and revenue.

```python
movies_counted.corr(method = 'spearman')
```

```python
#Plot the correlation between popularity and revenue
sns.regplot(x = 'popularity', y = 'revenue', data = movies_counted, line_kws = {'color': 'red'})
```

**Research Hypothesis (H):**

3. Highly budgeted movies return high profit.

```python
#Plot the correlation between budget and profit
sns.regplot(x = 'budget', y = 'profit', data = movies_counted, line_kws = {'color': 'red'})
```

**Research Hypothesis (H):**

4. Highly budgeted movies have a high popularity.

```python
#Add in release year
movies_genre = df[['popularity','budget','revenue','original_title','runtime','genres','release_year','vote_count','vote_average','profit']]
#Splitting the genres up per data entry
split = movies_genre['genres'].str.split('|').apply(Series,1).stack()
split.index = split.index.droplevel(-1)
split.name = 'genres_split'
del movies_genre['genres']
movies_genre = movies_genre.join(split)

movies_genre.head()
```

```python
#Group by year and genres and find the average
time_genre = pd.DataFrame(movies_genre.groupby(['release_year', 'genres_split'])['profit'].mean())
time_genre
```

```python
#Pivot date to have years as columns and genres as index
final_genre = pd.pivot_table(time_genre, values = 'profit', index = ['genres_split'], columns = ['release_year'])
```

```python
sns.set(rc = {'figure.figsize': (15,10)})
sns.heatmap(final_genre, cmap = 'YlGnBu', linewidths = .5)
plt.title('Genres by Profit per Year')
```
